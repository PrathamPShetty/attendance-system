```python
from mtcnn import MTCNN
import cv2
from tqdm import tqdm
from PIL import Image
import numpy as np
import os

# Initialize face detector
mtcnn = MTCNN()

# Load video
v_cap = cv2.VideoCapture('./test.mp4')
v_len = int(v_cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Create directory for saving faces
output_dir = 'D:/mtcnn/test'
os.makedirs(output_dir, exist_ok=True)

# Loop through video frames
for i in tqdm(range(v_len)):
    
    # Grab frame at intervals (e.g., every 50th frame)
    if i % 50 == 0:
        success, frame = v_cap.read()
        if not success:
            continue
        
        # Convert frame to RGB for MTCNN
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        pil_image = Image.fromarray(frame_rgb)
        
        # Detect faces
        faces = mtcnn.detect_faces(np.array(pil_image))
        
        # Save each detected face
        for j, face in enumerate(faces):
            x, y, w, h = face['box']
            extracted_face = frame_rgb[y:y+h, x:x+w]
            face_path = os.path.join(output_dir, f'face_{i}_{j}.jpg')
            cv2.imwrite(face_path, cv2.cvtColor(extracted_face, cv2.COLOR_RGB2BGR))
            print(f"Face {j} saved at {face_path}")

# Release the video capture object
v_cap.release()
```

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

# Load the video
v_cap = cv2.VideoCapture('/test.mp4')
v_len = int(v_cap.get(cv2.CAP_PROP_FRAME_COUNT))  # Define v_len

# Initialize variables
batch_size = 32
frames = []
boxes = []
landmarks = []
saved_frames = []

for idx in tqdm(range(v_len)):
    # Load frame
    success, frame = v_cap.read()
    if not success:
        continue

    # Process frame
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    pil_frame = Image.fromarray(frame_rgb)
    pil_frame = pil_frame.resize([int(f * 0.25) for f in pil_frame.size])  # Resize for faster processing
    frames.append(pil_frame)

    # When batch is full, detect faces and reset batch list
    if len(frames) >= batch_size:
        batch_boxes, _, batch_landmarks = mtcnn.detect(frames, landmarks=True)
        
        # Save each frame in the batch that contains detected faces
        for i, (box, landmark) in enumerate(zip(batch_boxes, batch_landmarks)):
            if box is not None and landmark is not None:
                # Save frame with faces detected
                frame_with_face = frames[i]
                save_path = f'detected_frame_{idx - batch_size + i + 1}.jpg'
                frame_with_face.save(save_path)
                saved_frames.append(frame_with_face)

        # Clear frames for the next batch
        frames = []

# Plot a subset of saved frames with detected boxes and landmarks
fig, ax = plt.subplots(3, 3, figsize=(18, 12))
for i in range(min(9, len(saved_frames))):  # Only plot up to 9 frames
    ax[int(i / 3), i % 3].imshow(saved_frames[i])
    ax[int(i / 3), i % 3].axis('off')
    
plt.show()
```
